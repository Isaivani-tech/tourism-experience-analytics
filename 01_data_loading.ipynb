```python
import pandas as pd
import os

# Set your data path
DATA_PATH = "data/"  # adjust if needed

# Load all datasets
transaction = pd.read_excel(DATA_PATH + "Transaction.xlsx")
user        = pd.read_excel(DATA_PATH + "User.xlsx")
city        = pd.read_excel(DATA_PATH + "City.xlsx")
country     = pd.read_excel(DATA_PATH + "Country.xlsx")
region      = pd.read_excel(DATA_PATH + "Region.xlsx")
continent   = pd.read_excel(DATA_PATH + "Continent.xlsx")
item        = pd.read_excel(DATA_PATH + "Item.xlsx")
type_df     = pd.read_excel(DATA_PATH + "Type.xlsx")
mode        = pd.read_excel(DATA_PATH + "Mode.xlsx")

# Preview each file
datasets = {
    "Transaction": transaction,
    "User": user,
    "City": city,
    "Country": country,
    "Region": region,
    "Continent": continent,
    "Item": item,
    "Type": type_df,
    "Mode": mode
}

for name, df in datasets.items():
    print(f"\n{'='*40}")
    print(f"📄 {name} — Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    print(df.head(3))
```

```python
# ============================================
# DATA CLEANING
# ============================================

print("=" * 50)
print("CHECKING MISSING VALUES")
print("=" * 50)

for name, df in datasets.items():
    missing = df.isnull().sum().sum()
    print(f"{name:15} → Missing values: {missing}")

print("\n" + "=" * 50)
print("CHECKING DUPLICATES")
print("=" * 50)

for name, df in datasets.items():
    dupes = df.duplicated().sum()
    print(f"{name:15} → Duplicate rows: {dupes}")

print("\n" + "=" * 50)
print("CHECKING DATA TYPES - TRANSACTION")
print("=" * 50)
print(transaction.dtypes)

print("\n" + "=" * 50)
print("TRANSACTION - Basic Stats")
print("=" * 50)
print(transaction.describe())

print("\n" + "=" * 50)
print("RATING VALUE COUNTS")
print("=" * 50)
print(transaction['Rating'].value_counts().sort_index())

print("\n" + "=" * 50)
print("VISIT MODE VALUE COUNTS")
print("=" * 50)
print(transaction['VisitMode'].value_counts())

print("\n" + "=" * 50)
print("USER - Missing CityId check")
print("=" * 50)
print(user.isnull().sum())
```

```python
# ============================================
# FIX MISSING VALUES & CLEAN DATA
# ============================================

print(f"User shape BEFORE cleaning: {user.shape}")
print(f"City shape BEFORE cleaning: {city.shape}")

# Drop rows with missing CityId in User
user = user.dropna(subset=['CityId'])

# Drop rows with missing values in City
city = city.dropna()

print(f"\nUser shape AFTER cleaning: {user.shape}")
print(f"City shape AFTER cleaning: {city.shape}")

# Convert CityId to int in User (it was float due to missing values)
user['CityId'] = user['CityId'].astype(int)

print("\n✅ Missing values fixed!")

# Verify Rating range (should be 1-5)
print(f"\nRating min: {transaction['Rating'].min()}")
print(f"Rating max: {transaction['Rating'].max()}")
print(f"Unique Ratings: {sorted(transaction['Rating'].unique())}")

# Verify VisitMode range
print(f"\nUnique VisitMode values: {sorted(transaction['VisitMode'].unique())}")

print("\n✅ Data Cleaning Complete!")
print(f"\nFinal shapes:")
print(f"  Transaction : {transaction.shape}")
print(f"  User        : {user.shape}")
print(f"  City        : {city.shape}")
```

```python
# ============================================
#  FIX SETTINGWITHCOPYWARNING
# ============================================

# Reload clean copies properly
user = user.copy()
user['CityId'] = user['CityId'].astype(int)

city = city.copy()

# Also reset index after dropping rows
user = user.reset_index(drop=True)
city = city.reset_index(drop=True)

print("✅ Warning fixed!")
print(f"User dtypes:\n{user.dtypes}")
print(f"\nCity dtypes:\n{city.dtypes}")
print(f"\nUser CityId sample: {user['CityId'].head().tolist()}")
```

```python
# ============================================
#  FINAL CLEAN MERGE - MASTER DATASET
# ============================================

# Start fresh
master = transaction.copy()
print(f"Start - Transaction: {master.shape}")

# Step 4a: + User
master = master.merge(user, on='UserId', how='left')
print(f"After + User: {master.shape}")

# Step 4b: + City — drop CountryId (duplicate)
city_clean = city.drop(columns=['CountryId'])
master = master.merge(city_clean, on='CityId', how='left')
print(f"After + City: {master.shape}")

# Step 4c: + Country — drop RegionId (duplicate)
country_clean = country.drop(columns=['RegionId'])
master = master.merge(country_clean, on='CountryId', how='left')
print(f"After + Country: {master.shape}")

# Step 4d: + Region — drop ContinentId (duplicate)
region_clean = region.drop(columns=['ContinentId'])
master = master.merge(region_clean, on='RegionId', how='left')
print(f"After + Region: {master.shape}")

# Step 4e: + Continent
master = master.merge(continent, on='ContinentId', how='left')
print(f"After + Continent: {master.shape}")

# Step 4f: + Item/Attraction
master = master.merge(item, on='AttractionId', how='left')
print(f"After + Item: {master.shape}")

# Step 4g: + Type
master = master.merge(type_df, on='AttractionTypeId', how='left')
print(f"After + Type: {master.shape}")

# Step 4h: + Mode
master = master.merge(mode, left_on='VisitMode', 
                      right_on='VisitModeId', how='left')
print(f"After + Mode: {master.shape}")

# ============================================
# CLEAN UP DUPLICATE/REDUNDANT COLUMNS
# ============================================
# Drop redundant ID columns we no longer need
cols_to_drop = ['VisitModeId', 'AttractionTypeId', 
                'AttractionCityId', 'AttractionAddress']
master = master.drop(columns=[c for c in cols_to_drop 
                               if c in master.columns])

# Rename VisitMode_y to VisitModeName for clarity
master = master.rename(columns={
    'VisitMode': 'VisitModeId',
    'VisitMode_y': 'VisitModeName'
})

# ============================================
# FINAL CHECK
# ============================================
print(f"\n{'='*50}")
print(f"✅ MASTER DATASET READY!")
print(f"Final Shape: {master.shape}")
print(f"\nAll Columns:\n{master.columns.tolist()}")
print(f"\nMissing values:\n{master.isnull().sum()}")
print(f"\nSample:")
print(master.head(3))
```

```python
# ============================================
# FINAL CLEANUP
# ============================================

# Rename VisitMode_x to VisitModeId
master = master.rename(columns={'VisitMode_x': 'VisitModeId'})

# Final column check
print("✅ Final Columns:")
print(master.columns.tolist())

print(f"\n✅ Final Shape: {master.shape}")

print("\n✅ Sample Data:")
print(master.head(3).to_string())

print("\n✅ Missing Values:")
print(master.isnull().sum())

# Save master dataset to CSV
master.to_csv('data/master_dataset.csv', index=False)
print("\n✅ Master dataset saved to data/master_dataset.csv")
```
